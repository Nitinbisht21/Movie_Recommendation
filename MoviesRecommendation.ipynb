```python
import pandas as pd
import numpy as np
```

```python
import ast
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies = movies.merge(credits , on="title")
```

```python
movies.info()
```

```python
movies = movies [['movie_id','genres','title','overview','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies ['keywords'] = movies ['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter +=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_dir(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['tags'] = movies["overview"]+ movies['keywords'] + movies["genres"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
new_df = movies[["movie_id","title","tags"]]
```

```python
new_df
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df["tags"][0]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
new_df.head()
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
list(similarity[0])
```

```python
sort=sorted([(i, float(v)) for i ,v in enumerate(similarity[0])],reverse=True,key=lambda x: x[1])[1:6]
```

```python
def recommend(movies):
  movies_index = new_df[new_df['title'] == movies].index[0]
  distance = similarity[movies_index]
  movies_list = sorted([(i, float(v)) for i ,v in enumerate(distance)],reverse=True,key=lambda x: x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

## Save model files for the Streamlit app

The frontend app loads `model/new_df.pkl` and `model/similarity.pkl`. Run the next cell after training the model in this notebook.

```python
import os
import pickle

os.makedirs("model", exist_ok=True)

# Keep the same title/tag data used for similarity, plus display fields for the frontend.
model_df = new_df.copy()
model_df["genres"] = movies["genres"].values
model_df["overview"] = movies["overview"].apply(lambda x: " ".join(x) if isinstance(x, list) else x).values
model_df["cast"] = movies["cast"].values
model_df["crew"] = movies["crew"].values
model_df = model_df.reset_index(drop=True)

with open("model/new_df.pkl", "wb") as f:
    pickle.dump(model_df, f)

with open("model/similarity.pkl", "wb") as f:
    pickle.dump(similarity, f)

print("Model files saved successfully.")
print(f"Movies: {len(model_df)}")
print(f"Similarity matrix: {similarity.shape}")
```
